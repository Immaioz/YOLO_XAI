```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to the "subset" folder
subset_folder = 'utils/dataset/subset_VIS'
labels = ['Ferry Boat', 'Speed Boat', 'Kayak', 'Sail Boat', 'Small Motor Boat', 'Yatch']
# List of subfolder names inside "subset"
subfolders = os.listdir(subset_folder)

# Create a Matplotlib figure to display the images
plt.figure(figsize=(12, 8))

# Loop through each subfolder
for i, subfolder in enumerate(subfolders):
    
    x = subfolder.split("_")
    # Get a list of image files in the subfolder
    image_files = [f for f in os.listdir(os.path.join(subset_folder, subfolder)) if f.endswith('.jpg')]
    # Select a random image file from the subfolder
    random_image = random.choice(image_files)

    # Load and display the random image
    img = mpimg.imread(os.path.join(subset_folder, subfolder, random_image))
    plt.subplot(2, 3, i + 1)  # Create a subplot for each image
    plt.imshow(img)
    plt.title(x[1])
    plt.axis("off")
plt.tight_layout()
plt.show()

```
